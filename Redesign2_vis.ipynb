## Redesign 2: Relações socioeconômicas e porcentagem de pessoas que respondem a perguntas em plataformas de Q&A

```python
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('classic')
import numpy as np

#plt.figure(figsize=(15,6))
```

```python
def plot_vis1_and_vis2(path, platform_name):
    plt.figure(figsize=(15,6))
    data_frame = pd.read_csv(path)

    perc_a, IDV = data_frame['perc_a'], data_frame['GNI']
    EPI, internet = data_frame['EPI'], data_frame['Internet']
    plt.scatter(IDV, perc_a, label=None,
                c=EPI, cmap='Blues',
                s=internet, linewidth=0, alpha=1)
    plt.axis(aspect='equal')
    plt.xlabel('GNI')
    plt.ylabel('Percentual de pessoas que responderam alguma questão')
    plt.colorbar(label='Proficiência Inglês')
    plt.clim(38, 70)

    for inter in [1, 55, 100]:
        plt.scatter([], [], c='k', alpha=0.3, s=inter,
                    label=str(inter))
    plt.legend(scatterpoints=1, frameon=False, labelspacing=1, title='Penetração Internet', loc=0)

    plt.title('Relação entre Renda Nacional Bruta (GNI) e Pessoas que respontem questões no ' + platform_name);
    plt.savefig(platform_name + ".png", bbox_inches='tight')
```

```python
df_stack = pd.read_csv('dados stackoverflow.csv')
df_super = pd.read_csv('dados superuser.csv')
```

```python
ax = df_stack.plot.scatter(x='IDV', y='perc_a', s=50, label='Stack Overflow', color='red');
plot_final = df_super.plot.scatter(x='IDV', y='perc_a', s=50, label='Super User', ax=ax, color='blue', figsize=(15,6));
plot_final.set_ylabel("Percentual de usuários que responderam alguma questão")
plot_final.set_xlabel("IDV")
```

Esta primeira visualização exibe, de forma geral, pontos relacionados ao Índice de individualidade (IDV) e o percentual de usuários que postaram alguma mensagem (StackOverflow e SuperUser). DIferente da visualização do paper, que exibe os países de forma ordenada pelo IDV, aqui exibe-se apenas os valores deste índice no enxo X, deixando a visualização mais leve, porém perdendo a informação dos países.

Os dados utilizados são:

-> IDV: Índice de individualismo de um país.

-> perc_a: porcentagem de usuários que responderam uma questão.

```python
plot_vis1_and_vis2("dados stackoverflow.csv", "Stack Overflow")
```

```python
plot_vis1_and_vis2("dados superuser.csv", "SuperUser")
```

As visualizações 2 e 3 exibem, de forma separada (StackOverflow e SuperUser), dados relativos ao porcentual de usuários que responderam ao menos uma questão nas plataformas (eixo Y) e o GNI (Eixo X). Também é considerado na vosualização, a penetração da Internet nos países (tamanho do círculo) e a proficiência em Inglês registrada nos países (tonalidade dos círculos).

Os dados utilizados foram:

-> perc_a: porcentagem de usuários que responderam uma questão;

-> GNI: Renda Nacional Bruta de um país;

-> Internet: Penetração da Internet em um país;

-> EPI: Proficiência em Inglês dos usuários das plataformas em um país;

Nestas visualizações, é possível perceber a correlação entre renda, acesso a Internet e proficiência em Inglês com o percentual de pessoas que reponderam ao menos uma pergunta. Isso provavelmente se deve ao fato de que estas três características (renda, acesso a Internet e proficiência em Inglês) aumenta as chances de uma dada pessoa ser usuário de uma das duas plataformas, que exigem acesso a um PC (por exemplo), Internet e, em sua marioria, estão escritos em Inglês.
